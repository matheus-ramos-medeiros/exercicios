```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Gráfico de Linha

df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv')
df['data'] = pd.to_datetime(df['data'])#colocando para ano
plt.plot('data','casos-confirmados', data=df)
plt.xlabel("Ano")
plt.ylabel("Casos Confirmados")
plt.title("Dengue em Campinas")
plt.show()
```

```python
# Gráfico de Barra
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/entrada_passageiro_linha1.csv")
df
# apagar linha TOTAL
df.drop(index=23, axis=0, inplace=True)
df.plot('Estação','Entradas',kind='bar')
plt.xlabel("Estação")
plt.ylabel("Entradas")
plt.title("Entradas na Estação")
plt.show()
```

```python
# Gráfico de Pizza

df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv")

# ver as primeiras entradas
df.head()

# extrair as ocorrências para um dicionário
df["ocorrencia_classificacao"].value_counts()

# preparar dados para o pie chart
labels = ['Incidente', 'Acidente', 'Incidente Grave']
valores = [2861, 1714, 633]

a = dict(df["ocorrencia_classificacao"].value_counts())

labels = a.keys()
valores = a.values()

# criando gráfico de pizza
plt.pie(x=valores, labels=labels)

# definindo título
plt.title("Classificação das Ocorrências")

# mostrando o gráfico
plt.show()
```

```python
# importar as bibliotecas necessárias
import matplotlib.pyplot as plt
from pywaffle import Waffle

# exemplo de Waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=4,
                columns=10,
                values={'Bolacha': 33,
                        'Biscoito': 48,
                        'Salgadinho': 19},
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)}
                )

fig.set_tight_layout(False)
plt.show()
```

```python
# importar o pandas
import pandas as pd

# importar dados do CENIPA
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv")

# extrair value counts da variavel ocorrencia_classificacao
data = dict(df["ocorrencia_classificacao"].value_counts())

# plotar gráfico de waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=5,
                columns=15,
                colors=("#232066", "#983D3D", "#DCB732"),
                values=data, icons='plane', icon_size=15,
                legend={'loc': 'upper left', 'bbox_to_anchor': (1.1,1)},
                icon_legend=True
                )
fig.set_tight_layout(False)
plt.show();
```

```python
# plotar homens e mulheres na sala de aula
fig = plt.figure(
    FigureClass=Waffle, 
    rows=10, columns=10, 
    values={'Homens': 100, 'Mulheres': 158}, 
    colors=("#232066", "#983D3D"),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    icons='child', icon_size=12, 
    icon_legend=True
)

fig.set_tight_layout(False)
plt.show();
```
